```python
#occasionally load time would spike to 1+ seconds, this is to try to remove outliers
def reject_outliers_2(data, m=10):
    d = numpy.abs(data - numpy.median(data))
    mdev = numpy.median(d)
    s = d / (mdev if mdev else 1.)
    return data[s < m]
```

```python
fullBit=Overlay('/home/xilinx/jupyter_notebooks/config_time/full_impl/cnvW1A2-ultra96.bit')
partialCnvw1a1='/home/xilinx/jupyter_notebooks/config_time/partials/cnvW1A1-ultra96.bit'
partialCnvw1a2='/home/xilinx/jupyter_notebooks/config_time/partials/cnvW1A2-ultra96.bit'
partialCnvw2a2='/home/xilinx/jupyter_notebooks/config_time/partials/cnvW2A2-ultra96.bit'
partialLfcw1a1='/home/xilinx/jupyter_notebooks/config_time/partials/lfcW1A1-ultra96.bit'
partialLfcw1a2='/home/xilinx/jupyter_notebooks/config_time/partials/lfcW1A2-ultra96.bit'
```

```python
runs = 100
#delay no longer used, was the initial attempt to prevent load time spikes by adding pauses during the loops
dly = 0.0
```

```python
#load full
x = 0
Tsum = 0
elapse = numpy.empty((1,runs))
for x in range (runs):
    start = time.time()
    fullBit.download()
    stop = time.time()
    time.sleep(dly)
    elapse[:,x] = (stop-start)
    time.sleep(dly)
print(elapse)
elapse = reject_outliers_2(elapse)
print(elapse)
fullBit_avg = numpy.average(elapse)
print(fullBit_avg)
```
